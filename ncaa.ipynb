import csv

# Define the filenames
input_filename = 'Fanatics-Product-Catalog_IR.txt'
output_filename = 'outputCollege2.csv'

# Define the number of rows to extract
num_rows = 1000000

# Read the first 1000 rows from the input file and write the filtered rows to the CSV file
try:
    with open(input_filename, 'r', encoding='utf-8') as infile, open(output_filename, 'w', newline='', encoding='utf-8') as outfile:
        reader = csv.reader(infile, delimiter='\t')
        writer = csv.writer(outfile)

        # Write header row
        header = next(reader)
        writer.writerow(header)

        row_count = 0

        for row in reader:
            if row[18] == "College":  # Assuming the "Category" column is the second column (index 1)
                writer.writerow(row)
                row_count += 1
                if row_count >= num_rows:
                    break

    print(f"First {num_rows} rows with 'Category' as 'College' have been written to {output_filename}.")
except FileNotFoundError:
    print(f"File {input_filename} not found.")

```python
import pandas as pd

df = pd.read_csv('outputCollege.csv')


```

```python
teams = [
     "Ohio State Buckeyes",
     "Michigan Wolverines",
     "Nittany Lions",
     "Crimson Tide",
     "Huskers",
     "Fighting Irish"
 ]

# exclude where Text3 contains any of the teams

df = df[~df['Text3'].str.contains('|'.join(teams))]


df['Text3'].value_counts()
```

```python

nf = df[df['Text1'].fillna('').str.contains("Sports & Outdoors~Fan Shop~Clothing & Accessories")]



# Count the occurrences of each value in the 'Text3' column of the 'nf' dataframe
# text3_counts = nf['Text3'].value_counts()

# Print the top 60 values
#  print(text3_counts.head(60))

nf = nf.loc[:, ['Unique Merchant SKU', 'Product Name', 'Product URL', 'Image URL', 'Current Price', 'Original Price', 'Product Description', 'Gender', 'Text1', 'Text2', 'Text3','Size']]

rename_dict = {
    'Unique Merchant SKU': 'ID',
    'Product Name': 'Title',
    'Product URL': 'Metafield: custom.product_url [url]',
    'Image URL': 'Image Src',
    'Current Price': 'Variant Price',
    'Original Price': 'Variant Compare At Price',
    'Product Description': 'Body HTML',
    'Gender': 'Metafield: custom.product_gender [single_line_text_field]',
    'Text1': 'Metafield: custom.text1 [single_line_text_field]',
    'Text2': 'Metafield: custom.text2 [single_line_text_field]',
    'Text3': 'Tags',
    'Size': 'Metafield: custom.sizes [single_line_text_field]'
}

nf = nf.rename(columns=rename_dict)



## remove emptyspaces in Tags column between values
nf['Tags'] = nf['Tags'].str.replace(' ', '')

# Create an additional column 'Variant Inventory Tracker' with all values value 'shopify'
nf['Variant Inventory Tracker'] = 'shopify'
nf['Variant Inventory Policy'] = 'continue'
nf['Variant Fulfillment Service'] = 'manual'
nf['Variant Inventory Qty'] = '1'




# # Split the DataFrame into multiple CSV files with less than 4,900 rows each
nf.to_csv(f'NCAA_Products_6teams.csv', index=False)

print("DataFrame saved to CSV files.")
# ## FILTERING NEW PRODUCTOS TO DISPLAY MOST SOLD JERSEYS
```
