```python
rename_dict = {
    'Unique Merchant SKU': 'ID',
    'Product Name': 'Title',
    'Product URL': 'Metafield: custom.product_url [url]',
    'Image URL': 'Image Src',
    'Current Price': 'Variant Price',
    'Original Price': 'Variant Compare At Price',
    'Product Description': 'Body HTML',
    'Gender': 'Metafield: custom.product_gender [single_line_text_field]',
    'Text1': 'Metafield: custom.text1 [single_line_text_field]',
    'Text2': 'Metafield: custom.text2 [single_line_text_field]',
    'Text3': 'Tags',
    'Size': 'Metafield: custom.sizes [single_line_text_field]'
}

nf = nf.rename(columns=rename_dict)
```

nf.head()

```python


# Extract unique tags
unique_tags = nf['Tags'].unique()

# Create the dictionary for renaming
rename_dict = {}

for tag in unique_tags:
    # Split the tag into words
    words = tag.split()
    # Take all words except the last one (assuming the last word is the team name)
    city_words = words[:-1]
    # Join the words to form the city name, lowercase it, and remove spaces
    city_name = ''.join(city_words).lower()
    # Add to the dictionary
    rename_dict[tag] = city_name

# Print the dictionary
print(rename_dict)

# Apply the renaming to the DataFrame
nf['Tags'] = nf['Tags'].map(rename_dict)

```

# Split the DataFrame into multiple CSV files with less than 4,900 rows each
chunk_size = 4900
num_chunks = len(nf) // chunk_size + 1

for i in range(num_chunks):
    start_row = i * chunk_size
    end_row = (i + 1) * chunk_size
    chunk_df = nf[start_row:end_row]
    chunk_df.to_csv(f'nf_Products_{i + 1}.csv', index=False)

print("DataFrames saved to CSV files.")
