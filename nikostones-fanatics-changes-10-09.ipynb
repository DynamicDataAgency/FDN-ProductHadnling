### Procesamiento para cargar a Shopify

```python

to_shopify_df = new_df.loc[:, ['Unique Merchant SKU', 'Product Name', 'Product URL', 'Image URL', 'Current Price', 'Original Price', 'Product Description', 'Gender', 'Text1', 'Text2', 'Text3','Size', 'Category']]

rename_dict = {
    'Unique Merchant SKU': 'ID',
    'Product Name': 'Title',
    'Product URL': 'Metafield: custom.product_url [url]',
    'Image URL': 'Image Src',
    'Current Price': 'Variant Price',
    'Original Price': 'Variant Compare At Price',
    'Product Description': 'Body HTML',
    'Gender': 'Metafield: custom.product_gender [single_line_text_field]',
    'Text1': 'Metafield: custom.text1 [single_line_text_field]',
    'Text2': 'Metafield: custom.text2 [single_line_text_field]',
    'Text3': 'Metafield: custom.team [single_line_text_field]',
    'Size': 'Metafield: custom.sizes [single_line_text_field]',
    'Category': 'Metafield: custom.league [single_line_text_field]',
}

to_shopify_df = to_shopify_df.rename(columns=rename_dict)



## remove emptyspaces in Tags column between values
to_shopify_df['Tags'] =  ''

# Create an additional column 'Variant Inventory Tracker' with all values value 'shopify'
to_shopify_df['Variant Inventory Tracker'] = 'shopify'
to_shopify_df['Variant Inventory Policy'] = 'continue'
to_shopify_df['Variant Fulfillment Service'] = 'manual'
to_shopify_df['Variant Inventory Qty'] = '1'
to_shopify_df['Status'] = 'Draft'

#filter rows with price > 50
to_shopify_df = to_shopify_df[to_shopify_df['Variant Price'] > 50]

# print tipe of variant price column
to_shopify_df['Variant Price'].dtype


to_shopify_df.head(10)
```

```python
# leave only columns that are needed: ID,Metafield: custom.text1 [single_line_text_field,Metafield: custom.text2 [single_line_text_field],Metafield: custom.team [single_line_text_field],Metafield: custom.league [single_line_text_field], Tags
to_shopify_df = to_shopify_df.loc[:, ['ID', 'Metafield: custom.text1 [single_line_text_field]', 'Metafield: custom.text2 [single_line_text_field]', 'Metafield: custom.team [single_line_text_field]', 'Metafield: custom.league [single_line_text_field]', 'Tags','Metafield: custom.product_url [url]']]
```

```python
to_shopify_df.head(10)
```

```python
# save as CSV
to_shopify_df.to_csv("data/Products_Jewlery_for_Niko_StonesV2.csv", index=False)
```
